```python
import pandas as pd
```

# Clean Data


```python
df = pd.read_csv("../audio-features-and-lyrics-of-spotify-songs/spotify_songs.csv")
df = df.drop(
    columns=["track_album_id", "playlist_id"]
)  # drop columns that are not useful
df.head(10)
```

```python
display(df.describe())
display(df.info())
```

## Show Languages


```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(figsize=(20, 10))
ax = sns.countplot(x="language", data=df, order=df["language"].value_counts().index)
```

## Scope to English language


```python
df_english = df[df["language"] == "en"]

display(df_english["language"].value_counts())
```

```python
df_english.hist(bins=50, figsize=(20, 15))
plt.show()
```

## Cleaning


```python
df_english.isnull().sum()
```

## Show Genres


```python
fig, ax = plt.subplots(figsize=(20, 10))
ax = sns.countplot(
    x="playlist_genre",
    data=df_english,
    order=df_english["playlist_genre"].value_counts().index,
)
```
